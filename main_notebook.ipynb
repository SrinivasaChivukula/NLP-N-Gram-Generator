```python
import docx
import nltk
import random
from nltk.util import ngrams
from collections import Counter, defaultdict

nltk.download('punkt')
nltk.download('punkt_tab')

```

```python
# 1. READING AND PROCESSING THE DATA
def get_tokens_from_docx(file_path):
    doc = docx.Document(file_path)
    full_text = " ".join([para.text for para in doc.paragraphs])
    tokens = [w.lower() for w in nltk.word_tokenize(full_text) if w.isalpha()]
    return tokens

tokens = get_tokens_from_docx('5 papers related to blockchain.docx')
```

```python
# 2. DEFINING THE N-GRAM GENERATOR FUNCTIONS
def build_ngram_model(tokens, n):
    """Builds a probability-based model for n-grams."""
    model = defaultdict(Counter)
    for ngram in ngrams(tokens, n):
        prefix = ngram[:-1]
        target = ngram[-1]
        model[prefix][target] += 1
    
    probs = {}
    for prefix, counts in model.items():
        total = sum(counts.values())
        probs[prefix] = {word: count/total for word, count in counts.items()}
    return probs

```

```python
def generate_sentence(model, n, length=10):
    """Generates a 10-word sentence using the built model."""
    current_prefix = random.choice(list(model.keys()))
    sentence = list(current_prefix)
    
    while len(sentence) < length:
        if current_prefix in model:
            candidates = list(model[current_prefix].keys())
            weights = list(model[current_prefix].values())
            next_word = random.choices(candidates, weights=weights)[0]
            sentence.append(next_word)
            current_prefix = tuple(sentence[-(n-1):])
        else:
            break
    return " ".join(sentence[:10])
```

```python
# 3. GENERATING THE OUTPUT FOR BIGRAM, TRIGRAM, AND 4-GRAM
for n in [2, 3, 4]:
    print(f"\n--- {n}-gram Language Generator ---")
    model = build_ngram_model(tokens, n)
    for i in range(5):
        print(f"Sentence {i+1}: {generate_sentence(model, n, 10)}")
```
